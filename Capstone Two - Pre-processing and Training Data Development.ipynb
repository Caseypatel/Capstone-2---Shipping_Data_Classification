# Imports

```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Load the Data 

```python
df = pd.read_csv("C:/Users/keshi/Downloads/archive (1)/Output_file1/EDA.csv")
```

```python
df.head()
```

# Label Encoding

```python
df1 = df.copy()
df1
```

```python
df1 = pd.get_dummies(df1,columns=['Mode_of_Shipment','Product_importance','Gender'], drop_first=True)
df1
```

```python
lE = LabelEncoder()
df1['Warehouse_block']=lE.fit_transform(df1['Warehouse_block'])
df1
```

```python
x = df1.drop(['Reached.on.Time_Y.N','ID'], axis=1)
x
```

```python
y = df1['Reached.on.Time_Y.N']
y
```

# Splitting Data 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Scaling Data

```python
scaler=StandardScaler()
x_train_scaled =scaler.fit_transform(x_train)
x_train=pd.DataFrame(x_train_scaled,columns=x.columns)
x_train
```

```python
x_test_scaled = scaler.transform(x_test)
x_test=pd.DataFrame(x_test_scaled,columns=x.columns)
x_test
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_train)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_train, y_pred)
print(accuracy)
```

```python
y_pred = LR.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

# Conclusion

The importance of the products in a shipment can influence shipping decisions and potential discounts on shipping costs. 
Businesses must strike a balance between meeting customer expectations, maintaining profitability, and managing shipping 
expenses when determining discounts in the context of product shipments.

Shipping by sea is an efficient method for transporting large quantities of goods over long distances. It is commonly 
used in international trade, facilitating the movement of goods between continents and the transport of bulk cargo. 
Given the volume and distance involved, most products are shipped by sea. Efficient product shipment is crucial for 
maintaining customer satisfaction, as delays or mishandling can lead to unhappy customers.

The mode of transportation plays a significant role in delivery times. For long distances, air freight is faster than 
shipping by sea. However, factors such as weather, transportation issues, or customs delays can impact road transportation 
for short-distance deliveries. Therefore, flexibility is essential when setting delivery expectations.
