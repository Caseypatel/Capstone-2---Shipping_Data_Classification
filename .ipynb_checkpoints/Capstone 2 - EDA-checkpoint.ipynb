# Label Encoding

```python
df = shipping_data.copy()
```

```python
df =pd.get_dummies(df,columns=['Mode_of_Shipment','Product_importance','Gender'],drop_first=True)
```

```python
df.head()
```

```python
df['Warehouse_block']=LabelEncoder().fit_transform(df['Warehouse_block'])
```

```python
df
```

# Chnaging non numeric data to numeric 

```python
df['Mode_of_Shipment_Road'] = df['Mode_of_Shipment_Road'].map({True: 1, False: 0})
```

```python
df['Mode_of_Shipment_Ship'] = df['Mode_of_Shipment_Ship'].map({True: 1, False: 0})
```

```python
df['Product_importance_medium'] = df['Product_importance_medium'].map({True: 1, False: 0})
```

```python
df['Gender_M'] = df['Gender_M'].map({True: 1, False: 0})
```
